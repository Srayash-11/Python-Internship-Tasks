#  Task 1: Matrix Operations Tool
This professional tool uses the **NumPy** library to perform complex linear algebra operations.

###  How to use:
1.  **Configure:** Use the form below to select your operation and dimensions.
2.  **Run:** Click the **Play button** on the left of the form.
3.  **Input:** Enter your matrix values when prompted in the output area.

```python
# @title Matrix Operations Tool
# @markdown Select settings and click Play.
operation = "Multiplication" # @param ["Addition", "Subtraction", "Multiplication", "Transpose", "Determinant"]
rows = 2 # @param {type:"integer"}
cols = 2 # @param {type:"integer"}

import numpy as np

def get_matrix_from_user(r, c):
    """
    Prompts the user for matrix elements and returns a NumPy array.
    """
    print(f"\n--- Input for {r}x{c} Matrix ---")
    data = []
    for i in range(r):
        row = list(map(float, input(f"  Row {i+1} values (space-separated): ").split()))
        if len(row) != c:
            print(f"[ERROR] Expected {c} values for this row.")
            return None
        data.append(row)
    return np.array(data)

# Main Execution Logic
print(f"Initializing {operation}...")

if operation in ["Addition", "Subtraction", "Multiplication"]:
    matrix_a = get_matrix_from_user(rows, cols)
    # For multiplication, Matrix B rows must match Matrix A columns
    matrix_b = get_matrix_from_user(cols, rows) if operation == "Multiplication" else get_matrix_from_user(rows, cols)

    if matrix_a is not None and matrix_b is not None:
        if operation == "Addition": result = matrix_a + matrix_b
        elif operation == "Subtraction": result = matrix_a - matrix_b
        elif operation == "Multiplication": result = np.dot(matrix_a, matrix_b)
        print("\n[RESULT]:\n", result)

elif operation == "Transpose":
    target_matrix = get_matrix_from_user(rows, cols)
    if target_matrix is not None:
        print("\n[TRANSPOSED MATRIX]:\n", target_matrix.T)

elif operation == "Determinant":
    if rows == cols:
        square_matrix = get_matrix_from_user(rows, rows)
        if square_matrix is not None:
            determinant_value = np.linalg.det(square_matrix)
            print(f"\n[DETERMINANT]: {determinant_value:.2f}")
    else:
        print("[ERROR] Determinant calculations require a square matrix (Rows = Cols).")

print("\n" + "-"*30)
print("Process completed successfully.")
print("-"*30)
```
